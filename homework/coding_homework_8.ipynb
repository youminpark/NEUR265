```python
# Question 2: import packages
import numpy as np;
import matplotlib.pyplot as plt;
from scipy.stats import pearsonr;
```

```python
# Question 3: import indices
fear_traces = np.loadtxt('https://raw.githubusercontent.com/youminpark/NEUR265/main/fear_conditioning_traces_hw8.csv', delimiter = ',');
extinction_predictors = np.loadtxt('https://raw.githubusercontent.com/youminpark/NEUR265/main/extinction_predictors_hw8.csv', delimiter = ',');
shock_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/shock_index.csv', delimiter = ',').astype(int);
tone_index = np.loadtxt('https://raw.githubusercontent.com/hallockh/neur_265/main/data/tone_index.csv', delimiter = ',').astype(int);
```

```python
# Question 5: make new variables
conditioning_traces = fear_traces[0:10457,:];
extinction_traces = fear_traces[10458:,:];
```

```python
# Question 6: print array sizes
print(np.shape(conditioning_traces));
print(np.shape(extinction_traces));
```

```python
# Question 7: create time variables
fps = 15;
conditioning_interval = len(conditioning_traces)/fps;
extinction_interval = len(extinction_traces)/fps;

time_conditioning = np.linspace(0,697.13,10457);
time_extinction = np.linspace(0,947.33,14210);
```

```python
# Question 9: generate a subplot of the first and second ROIs
fig, ax = plt.subplots(1,2,figsize=(15,5));

ax[0].plot(time_conditioning, conditioning_traces[:,0], 'b', alpha = 0.5);
ax[0].plot(time_conditioning, conditioning_traces[:,1], 'r', alpha = 0.5);
ax[0].plot([shock_index[0], shock_index[0]], [0,1], 'k', lw=2) # Make a black line at shock onset
ax[0].plot([shock_index[1], shock_index[1]], [0,1], 'k', lw=2) # Make a black line at shock onset
ax[0].plot([shock_index[2], shock_index[2]], [0,1], 'k', lw=2) # Make a black line at shock onset
ax[0].plot([tone_index[0], tone_index[0]], [0,1], 'g', lw=2)
ax[0].plot([tone_index[1], tone_index[1]], [0,1], 'g', lw=2)
ax[0].plot([tone_index[2], tone_index[2]], [0,1], 'g', lw=2)
ax[0].set_xlabel('Time (seconds)');
ax[0].set_ylabel('Voltage [$\mu V$]');
ax[0].set_title('Conditioning Data');

ax[1].plot(time_extinction, extinction_traces[:,0], 'b', alpha = 0.5);
ax[1].plot(time_extinction, extinction_traces[:,1], 'r', alpha = 0.5);
ax[1].set_xlabel('Time (seconds)');
ax[1].set_ylabel('Voltage [$\mu V$]');
ax[1].set_title('Extinction Data');
```

```python
# Question 12: develop GLM
shock_array = np.zeros(10457)
for i in range(10457):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10457)
for i in range(10457):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

first_trace = conditioning_traces[:,0]

predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
predictors_conditioning['shock onset'] = shock_array
r_model3 = np.zeros(63)

predictors_conditioning['shock onset'] = shock_array
model3 = sm.GLM(first_trace, predictors_conditioning, family=Poisson())
model3_results = model3.fit() # Fit model to our data
b1 = model3_results.params
print('b1:')
print(b1)
```

```python
# Question 13: run the GLM
predictors_conditioning['shock onset'] = shock_array
r_model_conditioning = np.zeros(63)

for i in range(63):
  trace_temp = conditioning_traces[:,i]
  model3 = sm.GLM(trace_temp, predictors_conditioning, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r_index_conditioning = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = r_index_conditioning.statistic

plt.plot(np.sort(r_model_conditioning), np.linspace(0, 1, len(r_model_conditioning), endpoint=False))
plt.legend()
plt.xlabel('Correlation Between Model Prediction and Observed Trace')
plt.ylabel('Cumulative Proportion of ROIs')
```

```python
# Question 14: now again for the extinction data!
shock_array = np.zeros(14210)
for i in range(14210):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(14210)
for i in range(14210):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0

first_trace = extinction_traces[:,0]

predictors_extinction = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
r_model3 = np.zeros(63)

predictors_extinction['shock onset'] = shock_array
model3 = sm.GLM(first_trace, predictors_extinction, family=Poisson())
model3_results = model3.fit() # Fit model to our data

predictors_extinction['shock onset'] = shock_array
r_model_extinction = np.zeros(63)

for i in range(63):
  trace_temp = extinction_traces[:,i]
  model_extinction = sm.GLM(trace_temp, predictors_extinction, family=Poisson())
  model3_results = model3.fit()
  predicted_trace = model3_results.predict()
  r_index_extinction = pearsonr(trace_temp, predicted_trace)
  r_model_extinction[i] = r_index_extinction.statistic

plt.plot(np.sort(r_model_extinction), np.linspace(0, 1, len(r_model_extinction), endpoint=False))
plt.legend()
plt.xlabel('Correlation Between Model Prediction and Observed Trace')
plt.ylabel('Cumulative Proportion of ROIs')
```
