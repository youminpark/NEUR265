```python
# Question 2: import packages
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
# Question 3: import files
left1_url = 'https://raw.githubusercontent.com/youminpark/NEUR265/main/spikes_left_trial1.csv';
left2_url = 'https://raw.githubusercontent.com/youminpark/NEUR265/main/spikes_left_trial2.csv';
right1_url = 'https://raw.githubusercontent.com/youminpark/NEUR265/main/spikes_right_trial1.csv';
right2_url = 'https://raw.githubusercontent.com/youminpark/NEUR265/main/spikes_right_trial2.csv';

spikes_left_trial1 = np.loadtxt(left1_url, delimiter = ',');
spikes_left_trial2 = np.loadtxt(left2_url, delimiter = ',');
spikes_right_trial1 = np.loadtxt(right1_url, delimiter = ',');
spikes_right_trial2 = np.loadtxt(right2_url, delimiter = ',');
```

```python
# Question 4: print lengths of each array
print('Left Trial 1:', len(spikes_left_trial1));
print('Left Trial 2:', len(spikes_left_trial2));
print('Right Trial 1:', len(spikes_right_trial1));
print('Right Trial 2:', len(spikes_right_trial2));
```

```python
# Question 6: determine firing rates
interval = 30;
L1frequency = len(spikes_left_trial1)/interval;
L2frequency = len(spikes_left_trial2)/interval;
R1frequency = len(spikes_right_trial1)/interval;
R2frequency = len(spikes_right_trial2)/interval;

print('Left Trial 1 Frequency:', L1frequency);
print('Left Trial 2 Frequency:', L2frequency);
print('Right Trial 1 Frequency:', R1frequency);
print('Right Trial 2 Frequency:', R2frequency);
```

##### Question 7: interpret results of frequencies
* There is no consistent pattern demonstrated by the frequencies of the trials. The left trial frequency decreased while the right trial frequency increased with trial progression. The first trial frequencies across conditions had no relationship, nor did the second trial frequencies.

```python
# Question 8: make two subplots for each trial
fig, ax = plt.subplots(1,2,figsize=(15,5));
ax[0].plot(spikes_left_trial1, label = 'Left');
ax[0].plot(spikes_right_trial1, label = 'Right');
ax[0].set_xlabel('Index');
ax[0].set_ylabel('Voltage [$\mu V$]');
ax[0].set_title('Trial 1 Data');
ax[0].legend();

ax[1].plot(spikes_left_trial2, label = 'Left');
ax[1].plot(spikes_right_trial2, label = 'Right');
ax[1].set_xlabel('Index');
ax[1].set_ylabel('Voltage [$\mu V$]');
ax[1].set_title('Trial 2 Data');
ax[1].legend();
```

##### Question 9: interpret results from question 8
* For trial 1, the right data is more "bursty" than the left, as its voltage spikes much faster.
* For trial 2, the left data is more "bursty" than the right, as its voltage spikes much faster.


```python
# Question 10: calculate ISI arrays for each data set
L1_ISI = np.diff(spikes_left_trial1);
L2_ISI = np.diff(spikes_left_trial2);
R1_ISI = np.diff(spikes_right_trial1);
R2_ISI = np.diff(spikes_right_trial2);
```

```python
# Question 11: create histograms for each ISI array
bins = np.arange(0, 2, 0.01)  # Define the bins for the histogram

fig, ax = plt.subplots(2,2,figsize=(15,10));
ax[0,0].hist(L1_ISI, bins);
ax[0,0].set_xlabel('ISI [s]')              # ... label the x-axis
ax[0,0].set_ylabel('Counts')               # ... and the y-axis
ax[0,0].set_title('Left Trial 1 ISIs')             # ... give the plot a title

ax[0,1].hist(L2_ISI, bins);
ax[0,1].set_xlabel('ISI [s]')              # ... label the x-axis
ax[0,1].set_ylabel('Counts')               # ... and the y-axis
ax[0,1].set_title('Left Trial 2 ISIs')             # ... give the plot a title

ax[1,0].hist(R1_ISI, bins);
ax[1,0].set_xlabel('ISI [s]')              # ... label the x-axis
ax[1,0].set_ylabel('Counts')               # ... and the y-axis
ax[1,0].set_title('Right Trial 1 ISIs')             # ... give the plot a title

ax[1,1].hist(R2_ISI, bins);
ax[1,1].set_xlabel('ISI [s]')              # ... label the x-axis
ax[1,1].set_ylabel('Counts')               # ... and the y-axis
ax[1,1].set_title('Right Trial 2 ISIs')             # ... give the plot a title
```

##### Question 12: interpret the histograms
* The left trials have a greater range of ISIs compared to the right trials, though the difference is marginal
* All four data sets exhibit bursting behavior since there is a wide range of ISIs, which suggests they fire irregularly. This is best seen in right trial 1, which notably has the lowest mode.
* The mode of all 4 trials was approximately 0.05-1 s.

```python
# Question 13: create two subplots of firing rates w.r.t. trial #
binsize = 1                                                                   # Let's start with 1-second bins
fr_bins = np.arange(0,30,binsize)                                             # Make time bins (in 1 second increments) from 0 to 30 s
L1fr_low = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array
L2fr_low = np.zeros(len(fr_bins)-1)
R1fr_low = np.zeros(len(fr_bins)-1)
R2fr_low = np.zeros(len(fr_bins)-1)

for i in range(len(fr_bins)-1):                                               # Loop through each bin
  L1spk_index = (spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1 < fr_bins[i+1])           # Find timestamp values falling within each bin
  L1true_values = spikes_left_trial1[L1spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  L1fr_low[i] = (len(L1true_values))/binsize

  L2spk_index = (spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2 < fr_bins[i+1])           # Find timestamp values falling within each bin
  L2true_values = spikes_left_trial2[L2spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  L2fr_low[i] = (len(L2true_values))/binsize

  R1spk_index = (spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1 < fr_bins[i+1])           # Find timestamp values falling within each bin
  R1true_values = spikes_right_trial1[R1spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  R1fr_low[i] = (len(R1true_values))/binsize

  R2spk_index = (spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2 < fr_bins[i+1])           # Find timestamp values falling within each bin
  R2true_values = spikes_right_trial2[R2spk_index]                                          # Our index is a boolean, so we need to isolate "True" values
  R2fr_low[i] = (len(R2true_values))/binsize

fig, ax = plt.subplots(1,2,figsize=(15,5));
ax[0].plot(fr_bins[:-1], L1fr_low, 'r');
ax[0].plot(fr_bins[:-1], R1fr_low, 'b');
ax[0].set_xlabel('Time [s]');              # ... label the x-axis
ax[0].set_ylabel('Firing Rate [Hz]');               # ... and the y-axis
ax[0].set_title('Trial 1 ISIs');             # ... give the plot a title

ax[1].plot(fr_bins[:-1], L2fr_low, 'r');
ax[1].plot(fr_bins[:-1], R2fr_low, 'b');
ax[1].set_xlabel('Time [s]');              # ... label the x-axis
ax[1].set_ylabel('Firing Rate [Hz]');               # ... and the y-axis
ax[1].set_title('Trial 2 ISIs');             # ... give the plot a title
```

##### Question 14: Interpret graphs from Question 13
* For trial 1, the left trials exhibit a higher firing rate overall, with significant peaks at 12 and 15 s respectively.
* For trial 2, the right trials exhibit a higher firing rate overall, with significant peaks at 2, 9, and 14 s.
* The firing pattern seems to change across trial number. Peaks in the left data occur during 10-20 s for Trial 1. In Trial 2, the peaks in the left data occur during 20-30 s.
* The peaks for the right data all generally occur before 15 s, though the exact time of their peaks do not coincide.
