```python
# Question 2: import packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Question 3: import data
hits_url = 'https://raw.githubusercontent.com/youminpark/NEUR265/main/acc_hits_hw5.csv';
fa_url = 'https://raw.githubusercontent.com/youminpark/NEUR265/main/acc_fa_hw5.csv';

acc_hits = np.loadtxt(hits_url, delimiter = ',');
acc_fa = np.loadtxt(fa_url, delimiter = ',');
```

```python
# Question 4: print the shapes
print(np.shape(acc_hits));
print(np.shape(acc_fa));
```

```python
# Question 6: create time variable
time = np.linspace(0,4,8001);
```

```python
# Question 7: find sampling frequency
frequency = (len(time)-1)/4
frequency
```

```python
# Question 8: generate subplots of data for first trial
fig, ax = plt.subplots(1,2,figsize=(15,5));
ax[0].plot(time, acc_hits[0]);
ax[0].set_xlabel('Time (s)');
ax[0].set_ylabel('Voltage [$\mu V$]')
ax[0].set_title('Trial 1 Hits')

ax[1].plot(time, acc_fa[0]);
ax[1].set_xlabel('Time (s)');
ax[1].set_ylabel('Voltage [$\mu V$]')
ax[1].set_title('Trial 1 False Alarms')
```

#### Question 9: describe the dominant oscillation and mention differences
* The dominant oscillation in the hits data is approx. 4 Hz.
* The dominant oscillation in the false alarms data is approx 1 Hz.
* The dominant oscillation for the false alarms is lower than that of the hits.

```python
# Question 10: calculate and plot hits ERP
plt.plot(time, acc_hits.mean(0));
plt.plot([2, 2], [-40,40], 'k', lw=2);
plt.plot([0, 4], [0,0], 'k', lw=1);
plt.xlabel('Time (s)');
plt.ylabel('Voltage [$\mu V$]')
plt.title('Hits ERP')
```

#### Question 11: interpret ERP results
* A wave of low frequency and high amplitude is observed after the screen-touch.

```python
# Question 12: calculate and plot false alarms ERP
plt.plot(time, acc_hits.mean(0), label = 'Hits');
plt.plot(time, acc_fa.mean(0), label = 'False Alarms');
plt.plot([2, 2], [-40,40], 'k', lw=2);
plt.plot([0, 4], [0,0], 'k', lw=1);
plt.xlabel('Time (s)');
plt.ylabel('Voltage [$\mu V$]')
plt.title('Hits and False Alarm ERP')
plt.legend();
```

#### Question 13: Interpret results of the combined plot
* A wave of low frequency and high amplitude is observed after the screen-touch for hits.
* There is no significant difference in response before and after the screen-touch in the false alarms data.
* The hits data displays a response after the stimulus while the false alarms does not.
